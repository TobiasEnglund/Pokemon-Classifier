## Result

The Torch augmentations resulted in lower accuracy than baseline. Albumentations performed a lot better, so we'll go with that.
I will go with **HueSaturationValue + RandomBrightnessContrast + ColorJitter + GaussNoise**.

- **HueSaturationValue, RandomBrightnessContrast, and ColorJitter** got top accuracy and suit well for Pok√©mon images, which have a lot of color variations.  
- **GaussNoise** adds robustness to the model and helps avoid overfitting.
